# Determine how to predict if a tumor is benign
Use PCA to analyze cancer data and determine the most important features to predict whether a tumor is benign

```python
import pandas as pd
import matplotlib.pyplot as pls
import numpy as np
import seaborn as sns
%matplotlib inline
```

### Import and explore data

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df.head()
```

```python
# Scale the data so all features have a single unit variance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled = scaler.transform(df)

# Determine the 2 most important features
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled)

# Transform data to its first principal components
x_pca = pca.transform(scaled)

# Plot first 2 principal components
sns.scatterplot(x_pca[:,0], x_pca[:,1], hue=cancer['target'])
```

There is a pretty clear separation between malignant and benign just based on the first two components! PCA is a "compression" algorithm of sorts, in that we can already get a pretty good idea of how to predict the state of a tumor based on 2 of the 30 original features!

These components aren't easy to make sense of, however. They don't map 1:1 to features in the dataset, but instead to combinations of features in the data.

```python
# Investigate the combination of features making up the first 2 principal components
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=df.columns)
df_comp
```

```python
sns.heatmap(df_comp, cmap='plasma')
```

We can now pass df_comp (with the principal components) into a logistic regression or SVM to predict malignant/benign for tumors! Looking at the graph below, we can see that a line would pretty nicely separate the two groupings of points, which represent malignant and benign tumors.

```python
sns.scatterplot(x_pca[:,0], x_pca[:,1], hue=cancer['target'])
```
