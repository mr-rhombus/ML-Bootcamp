# Random dataset
Generate a random set of points, then use k-means clustering to group points together

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Generate artificial data

```python
from sklearn.datasets import make_blobs
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=70)

# Plot blobs
x = data[0][:,0]
y = data[0][:,1]
color = data[1]

sns.scatterplot(x, y, hue=color)
```

### Create model

```python
from sklearn.cluster import KMeans

# Fit model
model = KMeans(n_clusters=4)
model.fit(data[0])

# See how model did
print('Cluster centers: \n', model.cluster_centers_, '\n')
print('Labels: \n', model.labels_)
```

Normally, our analysis would be done now, since we wouldn't have real labels to compare against. However, since we know the labels, we can compare to see how well the model did.

```python
# Evaluate metrics
from sklearn.metrics import classification_report, confusion_matrix
print('Classification report: \n', classification_report(data[1], model.labels_))
print('Confusino matrix: \n', confusion_matrix(data[1], model.labels_))
```
